# Librosa audio prossing

Pengolahan data mencari kata-kata paling keren menggunakan `librosa`

```python
import numpy as np
import librosa
import librosa.feature
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
from sklearn.preprocessing import minmax_scale as normalize
```

```python
y, sr = librosa.load('sample/sample.wav')
# ipd.Audio(y, rate=sr)
```

```python
spectral = librosa.feature.spectral_bandwidth(y)[0]
librosa.display.waveplot(y, sr=sr, alpha=.4)
plt.plot(normalize(spectral))
plt.show()
```
